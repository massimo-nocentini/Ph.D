## Pascal triangle

```python
pascal_rec = Eq(m_symbol[n+1,k+1], m_symbol[n,k] + m_symbol[n,k+1])
pascal_rec
```

```python
pascal_Zrec = Eq(m_symbol[n+1,0], m_symbol[n,0])
pascal_Zrec
```

```python
dims = (10,10)
pascal = symbolic_matrix(dims, m_symbol)
pascal_unfolded = unfold_in_matrix(pascal, Arec=pascal_rec, Zrec=pascal_Zrec)
pascal_unfolded[0]
```

### ...as combination of two matrices

```python
two_splitted_pascal = unfold_in_matrix(pascal, pascal_rec, unfold_row_start_index=2)
two_splitted_pascal[0]
```

```python
apply_subs(two_splitted_pascal[0], {m_symbol[1,1]:m_symbol[0,0]})
```

```python
unfold_upper_chunk(two_splitted_pascal, pascal_rec, unfolding_rows=2)[0]
```

### ...as combination of three matrices

```python
three_splitted_pascal = unfold_in_matrix(pascal, pascal_rec, unfold_row_start_index=3)
three_splitted_pascal[0]
```

```python
apply_subs(three_splitted_pascal, {m_symbol[1,1]:m_symbol[0,0],
                                   m_symbol[2,1]:m_symbol[1,0] + m_symbol[0,0],
                                   m_symbol[2,2]:m_symbol[0,0],
                                  })
```

```python
unfold_upper_chunk(three_splitted_pascal, pascal_rec, unfolding_rows=3)[0]
```

### ...as combination of four matrices

```python
four_splitted_pascal = unfold_in_matrix(pascal, pascal_rec, unfold_row_start_index=4)
four_splitted_pascal[0]
```

```python
apply_subs(four_splitted_pascal, {m_symbol[1,1]:m_symbol[0,0],
                                   m_symbol[2,1]:m_symbol[1,0] + m_symbol[0,0],
                                   m_symbol[2,2]:m_symbol[0,0],
                                  m_symbol[3,1]:m_symbol[2,0]+m_symbol[0,0]+m_symbol[1,0],
                                   m_symbol[3,2]:m_symbol[0,0]+m_symbol[1,0] + m_symbol[0,0],
                                   m_symbol[3,3]:m_symbol[0,0],
                                  })
```

```python
unfold_upper_chunk(four_splitted_pascal, pascal_rec, unfolding_rows=4)
```

## Catalan triangle

```python
c_symbol = IndexedBase('c')
dims = (10,10)
catalan_rec = build_rec_from_gf((1/(1-t), t, 10), c_symbol)
catalan_rec
```

```python
catalan = symbolic_matrix(dims, c_symbol)
unfolded_catalan = unfold_in_matrix(catalan, catalan_rec)
unfolded_catalan[0]
```

### ...as combination of two matrices

```python
two_splitted_catalan = unfold_in_matrix(catalan, catalan_rec, unfold_row_start_index=2)
two_splitted_catalan[0]
```

```python
apply_subs(two_splitted_catalan[0], {c_symbol[1,1]:c_symbol[0,0]})
```

```python
clean_two_splitted_catalan = unfold_upper_chunk(two_splitted_catalan, catalan_rec, unfolding_rows=2)
clean_two_splitted_catalan[0]
```

```python
catalan_matrix_expansion = extract_inner_matrices(clean_two_splitted_catalan, unfolding_rows=2)
```

```python
catalan_matrix_expansion[c_symbol[0,0]]
```

```python
catalan_matrix_expansion[c_symbol[1,0]]
```

```python
inits_dependencies = entail_dependencies(clean_two_splitted_catalan, catalan_rec, unfolding_rows=2)
inits_dependencies
```

```python
check_matrix_expansion(unfolded_catalan, catalan_matrix_expansion, inits_dependencies)
```

```python
def catalan_gf(t): return (1-sqrt(1-4*t))/(2*t)
c_gf = catalan_gf(t)
c_gf, c_gf.series(n=10)
```

```python
catalan_prime = t*(catalan_gf(t))**3
catalan_prime.simplify(), catalan_prime.series(n=10)
```

```python
a_k = expand(1 + t*catalan_prime.series(n=10))
b_k = expand(t*catalan_gf(t).series(n=10))
a_k, b_k
```

```python
a_k_1 = (catalan_gf(t)*a_k).series(n=10)
a_k_1
```

```python
b_k_1 = (catalan_gf(t)*b_k).series(n=10)
b_k_1
```

### ...as combination of three matrices

```python
clean_three_splitted_catalan = unfold_upper_chunk(three_splitted_catalan, catalan_rec, unfolding_rows=3)
clean_three_splitted_catalan[0]
```

```python
catalan_matrix_expansion = extract_inner_matrices(clean_three_splitted_catalan, unfolding_rows=3)
```

```python
catalan_matrix_expansion[c_symbol[0,0]]
```

# Motzkin triangles with generalizations

```python
a, b, c = symbols('a b c')
motzkin_rec = Eq(m_symbol[n+1,k+1], a*m_symbol[n,k] + b*m_symbol[n,k+1] + c*m_symbol[n,k+2]).subs({a:1})
motzkin_rec
```

```python
dims = (10,10)
motzkin = symbolic_matrix(dims, m_symbol, None, None)
unfolded = unfold_in_matrix(motzkin, motzkin_rec, unfold_row_start_index=1)
```

```python
substed = unfolded.subs({m_symbol[0,0]:1, a:1})
#substed.applyfunc(lambda term: term.factor(m_symbol[1,0], m_symbol[1,1]))
substed.applyfunc(lambda term: term.factor())
```

```python
m_pascal = unfolded.subs({a:1, b:1, c:0, m_symbol[0,0]:1})
m_pascal
```

```python
m_pascal = unfolded.subs({a:1, b:0, c:1, m_symbol[0,0]:1})
m_pascal
```

```python
m_motzkin = unfolded.subs({a:1, b:1, c:1, m_symbol[0,0]:1})
m_motzkin
```

```python
m_catalan = unfolded.subs({a:1, b:2, c:1, m_symbol[0,0]:1})
m_catalan
```

```python
m_catalan = unfolded.subs({a:1, b:1, c:2, m_symbol[0,0]:1})
m_catalan
```

```python
m_motzkin = unfolded.subs({a:1, b:2, c:2, m_symbol[0,0]:1})
m_motzkin
```

### ...as combination of two matrices

```python
two_splitted_motzkin = unfold_in_matrix(motzkin, motzkin_rec, unfold_row_start_index=2)
clean_two_splitted_motzkin = unfold_upper_chunk(two_splitted_motzkin, motzkin_rec, unfolding_rows=2)
clean_two_splitted_motzkin = clean_two_splitted_motzkin.applyfunc(
    lambda term: Poly(term, [m_symbol[0,0], m_symbol[1,0]]).args[0])
clean_two_splitted_motzkin
```

```python
motzkin_matrix_expansion = extract_inner_matrices(clean_two_splitted_motzkin, m_symbol, unfolding_rows=2)
```

```python
motzkin_matrix_expansion[m_symbol[0,0]]
```

```python
motzkin_matrix_expansion[m_symbol[1,0]]
```

```python
check_matrix_expansion(unfolded, motzkin_matrix_expansion, {m_symbol[1,0]:b*m_symbol[0,0]})
```

```python
a_symbol = Function('a')
order = 10
spec = a_symbol, t, order
A_sequence = make_abstract_A_sequence(spec)
A_sequence
```

```python
A_sequence.args
```

```python
g_symbol = IndexedBase('g')
dims = (order,order)
generic_rec = build_rec_from_gf((A_sequence, t, order), g_symbol)
generic_rec
```

```python
generic_rec.rhs.args
```

```python
dims = (order,order)
generic_matrix = symbolic_matrix(dims, g_symbol, None, None)
unfolded_generic = unfold_in_matrix(generic_matrix, generic_rec, unfold_row_start_index=1)
```
